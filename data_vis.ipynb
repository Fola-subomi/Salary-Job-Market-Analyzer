```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import plotly.express as px
import csv
```

```python
df = pd.read_csv('data/clean/jobs_clean.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df_salary = df.dropna(subset=['avg_salary'])
```

```python
df['date_posted'] = df['date_posted'].fillna('Unknown Date')
```

```python
print(f"Total rows before: {len(df)}")
print(df[['date_posted', 'avg_salary']].isnull().sum())
```

```python
df.head()
```

```python
# Creating the Bar Chart
# Sort by salary descending
df_sorted = df.sort_values('avg_salary', ascending=False)

fig_bar = px.bar(df_sorted, 
                 x=df['title'], 
                 y=df['avg_salary'],
                 color=df['avg_salary'],  # Color gradient based on value
                 text_auto='.2s',     # Adds labels on top of bars
                 title="Average Salary by Job Title",
                 labels={'Salary_Avg': 'Average Salary ($)', 'Job_Title': 'Role'},
                 template='plotly_white')

fig_bar.show()
```

```python
df_avg_salary = df['avg_salary'].fillna('Confidential')
print(df_avg_salary.isnull.sum())
```

```python
fig = px.treemap(
    df, 
    path=[px.Constant("Job Market"), 'sector', 'title'], 
    values=df['location'],
    color=df['avg_salary'],
    color_continuous_scale='RdBu',
    title='Job Market Analysis: Industry Volume vs. Average Salary'
)

# 3. Enhance formatting
fig.update_traces(textinfo="label+value")
fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

# 4. Display
fig.show()
```
